## **1. Import Libraries and Load Datasets**

You will find the datasets in the folder titled Tutorial 3 in LMS. Download the data then upload it here in the file browser in the left bar.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
movies_df = pd.read_csv('Resources/movies.csv', delimiter = '|', on_bad_lines = 'skip')
genres_df = pd.read_csv('Resources/genre.csv', delimiter = '|')
directors_df = pd.read_csv('Resources/director.csv', delimiter = '|', on_bad_lines = 'skip')
```

## **2. Merge Datasets**

join movies.csv with genres.csv on id and movie_id and then join the result with directors.csv on dir_id.

```python
merged_df = pd.merge(movies_df, genres_df, left_on = 'id', right_on = 'movie_id')
merged_df = pd.merge(merged_df, directors_df, left_on = 'dir_id', right_on = 'id')
merged_df.head()
```
